```python
# from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x= df.drop("cls", axis = 1)
y=df["cls"]
```

```python
# Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# # Feature extraction: Convert text data to binary format using CountVectorizer.
# count_vectorizer = CountVectorizer(binary=True)
# X_train_binary = count_vectorizer.fit_transform(X_train)
# X_test_binary = count_vectorizer.transform(X_test)
```

```python
print(X_train.shape, y_train.shape)
```

```python
# Model training: Create and train a Bernoulli Naive Bayes classifier.
clf = BernoulliNB()
clf.fit(X_train, y_train)
```

```python
# Model prediction: Make predictions on the testing set.
y_pred = clf.predict(X_test)
```

```python


# Evaluation: Calculate accuracy.
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
